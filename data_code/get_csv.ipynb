```python
import pandas as pd
```

```python
url_marine = 'https://en.wikipedia.org/wiki/List_of_marine_aquarium_fish_species'
# get all the table from url
html_tables_marine = pd.read_html(url_marine, header=0)
```

```python
# concat the marine list
table_mar = pd.concat(html_tables_marine[:-1])
# -1 เพราะตอนท้ายมันมีตารางข้อมูลที่ไม่ใช่สปีชีส์ปลาอยู่

# drop unnecessary things
table_mar = table_mar.drop(["Image", "Reef safe","Care Level", "Temp.", "Description", "Max size"], axis=1)
table_mar.rename(columns = {'Taxonomy':'scientific'}, inplace = True)
table_mar.rename(columns = {'Common name':'common'}, inplace = True)
table_mar = table_mar.reset_index(drop=True)

table_mar
```

```python
# sort A-Z and reset index
table_mar = table_mar.sort_values(by=['scientific'])
table_mar = table_mar.reset_index(drop=True)
table_mar
```

```python
# save to csv
table_mar.to_csv('df.csv')
```
