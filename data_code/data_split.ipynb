```python
# get all classes name
path = 'dataset_no_group/data/'
files = os.listdir(path)
files
```

```python
# make train, test, valid dir
for i in files:
    os.makedirs(f'dataset_no_group/train/{i}')
    os.makedirs(f'dataset_no_group/valid/{i}')
    os.makedirs(f'dataset_no_group/test/{i}')
```

```python
# to check number of files in each class folder
if os.path.isdir('dataset_no_group/train/Abalistes_stellatus') is True:
  num = []
  for i in files:
    lst = os.listdir(f'dataset_no_group/train/{i}')
    number_files = len(lst)
    num.append(number_files)
```

```python
num
```

```python
# splitting to 70:20:10 and move splitted files to train, test, valid folder
for i in files:
  source_dir = f'dataset_no_group\data\{i}'
  split_files = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]

  train_files, test_valid_files = train_test_split(split_files, test_size=0.3, random_state=42)
  valid_files, test_files = train_test_split(test_valid_files, test_size=0.33, random_state=42)

  def move_files(file_list, dest_dir):
      os.makedirs(dest_dir, exist_ok=True)
      for file in file_list:
          shutil.move(os.path.join(source_dir, file), os.path.join(dest_dir, file))

  move_files(train_files, f'dataset_no_group\train\{i}')
  move_files(valid_files, f'dataset_no_group\valid\{i}')
  move_files(test_files, f'dataset_no_group\test\{i}')
```
